# 1ère série de demandes

## Requête: proportion de personnes en état de sous-nutrition

```python
# Lecture du dataset concernant la population mondiale
dfpop = pd.read_csv("population.csv")
```

```python
# Lecture du dataset concernant la population sous-alimentée
dfsn = pd.read_csv("sous_nutrition.csv")
```

```python
# Part des valeurs NaN dans le dataframe population en Sous-Nutrition (dfsn)
dfsn.isna().mean()
```

```python
# Remplacement des valeurs concernant la variable "Année" dans dfsn
vals_to_replace = {'2012-2014':'2013', '2013-2015':'2014', '2014-2016':'2015', '2015-2017': '2016', '2016-2018': '2017', '2017-2019': '2018'}
dfsn['Année'] = dfsn['Année'].map(vals_to_replace)
dfsn
```

```python
# Application de la méthode info() au dataframe "sous-nutrition"
dfsn.info()
```

```python
# Changement des valeurs de la variable "Année" en nombre entier. 
dfsn = dfsn.astype({ "Année":'int64'}) 
dfsn.dtypes
```

```python
# Filtrer pour l'année 2017 dataframe "sous-nutrition"
dfsn=dfsn[dfsn["Année"] == 2017]
```

```python
# Filtrer pour l'année 2017 dataframe "population"
dfpop=dfpop[dfpop["Année"] == 2017]
```

```python
# Jointure des dataframes "POPulation" et "Sous-Nutrition" pour créer un nouveau dataframe (dfpopsn)
dfpopsn = pd.merge(dfpop, dfsn, on='Zone', how='left')
dfpopsn
```

```python
# Ne pas prendre en compte valeurs manquantes des Valeurs concernant population sous-alimentées
dfpopsn[dfpopsn["Valeur_y"].notnull()]
```

```python
dfpopsn.info()
```

```python
# Supprimer une des variables Année ('Année_y'), redondante
dfpopsn=dfpopsn.drop(['Année_y'], axis = 1)
```

```python
# Remplacer valeurs <0.1 par 0
dfpopsn=dfpopsn.replace('<0.1', '0')
```

```python
# Changement de type de 'Valeur_y' représentant le nombre de millions de personnes en situation de sous-alimentation en float
dfpopsn = dfpopsn.astype({ "Valeur_y":'float64'}) 
dfpopsn.dtypes
```

```python
# Convertir population totale(Valeur_x) en millions d'habitants pour s'ajuster à l'unité de la population sous nourrie ('Valeur_y')
dfpopsn['Valeur_x']=dfpopsn['Valeur_x'] / 1000
```

```python
dfpopsn=dfpopsn.rename(columns = {'Valeur_x' : 'Population', 'Valeur_y' : 'Population sous-alimentées'})
```

```python
dfpopsn
```

```python
# Proportion de la population sous-alimentée
Proportion = (dfpopsn["Population sous-alimentées"].sum()) / (dfpopsn["Population"].sum()) * 100
print(Proportion)
```

##### Il y a environ 7,1% de la population mondiale en état de sous-nutrition.

## Requête: nombre théorique de personnes qui pourraient être nourries à partir de la disponibilité alimentaire mondiale

```python
dfpopsn
```

```python
# Lecture du dataset concernant la disponibilité alimentaire 
dfdispo = pd.read_csv("dispo_alimentaire.csv")
```

```python
# Regroupement des données en fonction de la variable 'Zone' + Agrégation (/somme) des valeurs de la variable 'Disponibilité alimentaire (Kcal/personne/jour)'
dfdispo=dfdispo.groupby(["Zone"])['Disponibilité alimentaire (Kcal/personne/jour)'].sum()
```

```python
# Jointure des dataframes sur la population sous-alimentée et la disponibilité alimentaire
df2=pd.merge(dfpopsn, dfdispo, on='Zone', how='left')
```

```python
# Renommer colonne 'Année_x' en 'Année'
df2=df2.rename(columns = {'Année_x' : 'Année'})
```

```python
df2
```

```python
# Ne pas prendre en compte valeurs manquantes de la colonne suivante
var_modif=['Disponibilité alimentaire (Kcal/personne/jour)']

for var in var_modif:
    df2=df2[df2[var].notnull()]
```

```python
df2
```

```python
# Calcul du nombre de personnes pouvant être nourries en théorie dans le monde selon la disponibilté alimentaire (en millions) 
NombreDePersonnesPouvantEtreNourriesEnTheorie = (df2["Disponibilité alimentaire (Kcal/personne/jour)"].mean()/2500) * df2["Population"].sum()
print(NombreDePersonnesPouvantEtreNourriesEnTheorie)
```

##### En théorie, il y a environ 8 288 381 473 personnes qui pourraient être nourries à partir de la disponibilité alimentaire mondiale

```python
# Proportion
round(NombreDePersonnesPouvantEtreNourriesEnTheorie / df2["Population"].sum(),1)
```

##### soit environ 1,1 fois la population totale

```python
dfdispoVeg
```

## Requête: utilisation de la disponibilité intérieure

```python
# Lecture du dataset concernant la disponibilité alimentaire
dfdispoInt=pd.read_csv("dispo_alimentaire.csv")
```

```python
# Regroupement des données en fonction de la variable 'Zone' + Agrégation (/somme) des valeurs les trois variables interrogées
dfdispoInt=dfdispoInt.groupby(["Zone"]).agg({'Disponibilité intérieure':'sum','Nourriture':'sum', 'Aliments pour animaux':'sum', 'Pertes':'sum'})
```

```python
dfdispoInt
```

```python
# Part de la disponibilité intérieure pour l'alimentation animale
PartDeLaDisponibilitéIntérieureAttribuéeAl_AlimentationAnimale=(dfdispoInt["Aliments pour animaux"].sum() / dfdispoInt["Disponibilité intérieure"].sum()) * 100
print(round(PartDeLaDisponibilitéIntérieureAttribuéeAl_AlimentationAnimale,2),"%")
```

```python
# Part de la disponibilité intérieure pour l'alimentation humaine
PartDeLaDisponibilitéIntérieureAttribuéePourl_AlimentationHumaine=(dfdispoInt["Nourriture"].sum() / dfdispoInt["Disponibilité intérieure"].sum()) * 100
print(round(PartDeLaDisponibilitéIntérieureAttribuéePourl_AlimentationHumaine,2),"%")
```

```python
# Part des pertes
PartDeLaDisponibilitéIntérieurePerdue=(dfdispoInt["Pertes"].sum() / dfdispoInt["Disponibilité intérieure"].sum()) * 100
print(round(PartDeLaDisponibilitéIntérieurePerdue,2),"%")
```

```python
dfdispoInt["Pertes"].sum()
```

```python
# Reste (en milliers de tonnes)
ResteDeLaDisponibilitéIntérieure=dfdispoInt["Disponibilité intérieure"].sum()*(1-(PartDeLaDisponibilitéIntérieureAttribuéeAl_AlimentationAnimale+PartDeLaDisponibilitéIntérieureAttribuéePourl_AlimentationHumaine+PartDeLaDisponibilitéIntérieurePerdue)/100)
print(ResteDeLaDisponibilitéIntérieure)
```

```python
# Part pour le reste de la disponibilité intérieure (Autres usages)
Reste=(ResteDeLaDisponibilitéIntérieure/dfdispoInt["Disponibilité intérieure"].sum()) *100
print(round(Reste,2),"%")
```

```python
# Construction d'un graphique en secteur sur l'usage de la disponibilité intérieure
x = [49.51,13.24,4.61,32.64]
plt.pie(x, labels = ['Nourriture', 'Alimentation animaux', 'Pertes', 'Autres'], normalize = True)
plt.figure(figsize = (10, 10))
plt.title("Utilisation de la disponibilité intérieure", fontsize = 40)


plt.pie(x, labels = ['Nourriture', 'Alimentation animaux', 'Pertes', 'Autres'], normalize = True)
plt.legend()
```

# Demandes subsidiaires

## Répartion de l'utilisation des céréales

```python
dfdispo = pd.read_csv("dispo_alimentaire.csv")
```

```python
# Filtrer dans la variable 'Produit' en fonction des valeurs composant la liste de céréales
dfdispo.query(' Produit == ["Blé et produits", "Riz et produits", "Orge et produits", "Maïs et produits", "Seigle et produits", "Avoine", "Millet et produits", "Sorgho et produits", "Céréales, Autres"]', inplace=True)
```

```python
# Supprimer les potentiels doublons
dfdispo= dfdispo.drop_duplicates()
```

```python
# Remplacer les valeurs manquantes par la valeur 0 dans le dataframe 
dfdispo=dfdispo.fillna(0)
```

```python
# Regroupement des données en fonction de la variable 'Produit' + Agrégation (/somme) des valeurs pour chacune des variables 'Aliments pour animaux', 'Nourriture', 'Disponibilitéintérieure' 
dfdispo=dfdispo.groupby(['Produit']).agg({"Aliments pour animaux" : "sum", "Nourriture" : "sum", "Disponibilité intérieure": "sum"})
```

```python
# Quantité (milliers de tonnes) de la disponibilité en céréales consacrée à l'alimentation pour les animaux
dfdispo['Aliments pour animaux'].sum()
```

```python
# Quantité de la disponibilité en céréales consacrée à l'alimentation humaine
dfdispo['Nourriture'].sum()
```

```python
# Quantité de la disponibilité composée de céréales
dfdispo['Disponibilité intérieure'].sum()
```

```python
# Calcul de la part des céréales utilisées dans l'alimentation pour les animaux
Part_Aliments_Animaux=dfdispo['Aliments pour animaux'].sum()/dfdispo['Disponibilité intérieure'].sum() *100
print(round(Part_Aliments_Animaux,2),"%")
```

```python
# Calcul de la part des céréales utilisées dans l'alimentation humaine
Part_Alimentation_Humains=dfdispo['Nourriture'].sum()/dfdispo['Disponibilité intérieure'].sum()*100
print(round(Part_Alimentation_Humains,2),"%")
```

```python
# Calcul de la quantité ( en milliers de tonnes) de céréales utilisées pour d'autres usages
Autres_Usages= dfdispo['Disponibilité intérieure'].sum()*(1-(Part_Aliments_Animaux+Part_Alimentation_Humains)/100)
round(Autres_Usages,2)
```

```python
# Calcul de la part des céréales utilisées dans pour d'autres usages
dfPart_Autres_Usages=(Autres_Usages / dfdispo['Disponibilité intérieure'].sum()) * 100
print(round(dfPart_Autres_Usages,1),"%")
```

```python
# Construction d'un graphique en secteur qui montre l'usage qui est fait des céréales
x = [69.34,18.13,12.5]
plt.pie(x, labels = ['Alimentation animaux','Nourriture','Autres Usages'], normalize = True)
plt.figure(figsize = (10, 10))
plt.title("Utilisation des céréales (répartition)", fontsize = 40)


plt.pie(x, labels = ['Alimentation animaux','Nourriture','Autres Usages'], normalize = True)
plt.legend()
```

A titre comparatif, nous y joindrons un cinquième qui du fait d'une très forte production, à l'inverse, fournit de l'aide alimentaire. De plus, nous allons intentionnellement choisir de ne pas prendre en compte certaines variables telles que les importations ou les exportations qui ont une influence sur la disponibilité pour ne rester concentré que sur son rapport à la production.
##### La question: Dans les cas choisis, la production d'un pays suffit elle à assurer la disponibilité apportée à sa population et la suffisance alimentaire? 

```python
dfDisponibilité = pd.read_csv("dispo_alimentaire.csv")
```

```python
# Regroupement des données en fonction de la variable 'Zone' + Agrégation (/somme) des valeurs pour chacune des variables 'Production', 'Disponibilité alimentaire en quantité (kg/personne/an)', Exportations - Quantité','Importations - Quantité' 
dfDisponibilité=dfDisponibilité.groupby(["Zone"]).agg({'Production' : 'sum','Disponibilité alimentaire en quantité (kg/personne/an)': "sum", 'Exportations - Quantité': 'sum', 'Importations - Quantité' : 'sum' })
dfDisponibilité
```

```python
dfPopulation = pd.read_csv("population.csv")
```

```python
# Filtrer dataframe population pour la seule année 2017
dfPopulation=dfPopulation[dfPopulation["Année"] == 2017]
```

```python
# Renommer la variable 'Valeur' en 'Population totale'
dfPopulation=dfPopulation.rename(columns = {'Valeur':'Population totale'})
```

```python
# Convertir la population totale en millions d'habitants
dfPopulation['Population totale']=dfPopulation['Population totale'] *1000
dfPopulation
```

```python
# Jointure du dataframe disponibilité avec le dataframe population pour obtenir un nouveau dataframe : Dispopulation
dfDispopulation=pd.merge( dfPopulation, dfDisponibilité, how='inner', on='Zone')
dfDispopulation
```

```python
# Créer une nouvelle variable affichant la disponibilité pour chacune des populations des pays, convertie en milliers de tonnes
dfDispopulation=dfDispopulation.assign(Disponibilité_Population=(dfDispopulation['Disponibilité alimentaire en quantité (kg/personne/an)']) * (dfDispopulation['Population totale']) /1000000 )
```

```python
dfDispopulation=round(dfDispopulation)
dfDispopulation
```

```python
# Assigner une (nouvelle) colonne pour la part de la production dans la disponibilité 
dfDispopulation= dfDispopulation.assign(Part_de_la_production_dans_la_disponibilité=(dfDispopulation["Production"] / dfDispopulation["Disponibilité_Population"])*100)
```
